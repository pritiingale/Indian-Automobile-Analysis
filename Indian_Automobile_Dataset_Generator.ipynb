### Import Libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
import random
```

### Set random seed

```python
# Set random seed (so data is reproducible)
np.random.seed(42)
```

### Number of Records

```python
# Decide number of records (rows)
records = 1000
```

### Create dictionary of car brands

```python
# Create dictionary of car brands and their models
brands_models = {
    "Maruti": ["Swift", "Baleno", "WagonR", "Alto"],
    "Hyundai": ["i10", "i20", "Creta", "Verna"],
    "Honda": ["City", "Amaze", "Civic"],
    "Tata": ["Nexon", "Tiago", "Altroz"],
    "Mahindra": ["XUV300", "Scorpio", "Bolero"],
    "Toyota": ["Innova", "Fortuner", "Glanza"]
}
```

### Define possible values for other columns

```python
# Define possible values for other columns
cities = ["Delhi", "Mumbai", "Bangalore", "Chennai", "Hyderabad", "Pune", "Kolkata"]
fuel_types = ["Petrol", "Diesel", "CNG", "Electric"]
transmissions = ["Manual", "Automatic"]
owners = ["First Owner", "Second Owner", "Third Owner"]
```

### List of empty data

```python
# Create empty list to store data
data = []
```

### Generate random data for each record

```python
# Generate random data for each record
for i in range(records):
    
    # Choose brand and model
    brand = random.choice(list(brands_models.keys()))
    model = random.choice(brands_models[brand])
    
    # Random year of manufacture
    year = random.randint(2005, 2024)
    
    # Random price (with 5% missing values)
    price = random.randint(80000, 2500000)
    if random.random() < 0.05:
        price = None  # introduce missing values
    
    # Other numeric features
    km = random.randint(5000, 200000)
    mileage = round(random.uniform(10, 30), 1)
    engine = random.choice([800, 1000, 1200, 1500, 1800, 2000])
    seats = random.choice([4, 5, 7])
    
    # Append the row into data list
    data.append([
        i + 1,
        brand,
        model,
        year,
        price,
        km,
        random.choice(fuel_types),
        random.choice(transmissions),
        random.choice(owners),
        mileage,
        engine,
        seats,
        random.choice(cities),
        pd.Timestamp("2019-01-01") + pd.to_timedelta(random.randint(0, 2000), unit="D")
    ])
```

### Define Column Names

```python
# Define column names
columns = [
    "car_id", "brand", "model", "year", "price", "km_driven",
    "fuel_type", "transmission", "owner_type", "mileage_kmpl",
    "engine_cc", "seats", "city", "listing_date"
]
```

### DataFrame

```python
# Create DataFrame
df = pd.DataFrame(data, columns=columns)
```

### **Missing Value Handling (price)**

```python
# Check missing values in each column
df.isnull().sum()
```

### Median of price

```python
# Calculate median price (robust to outliers)
median_price = df['price'].median()
```

```python
# Fill missing price values with median
df['price'] = df['price'].fillna(median_price)
```

### Checking there was no missing values left

```python
# Verify no missing values left
df.isnull().sum()
```

### **Checking Data Types of Dataset**

```python
# Check current data types
df.dtypes
```

### **Type conversion (listing_date)**

```python
# Convert listing_date from object to datetime format
df['listing_date'] = pd.to_datetime(df['listing_date'])
```

```python
# Confirm the change
df.dtypes
```

### **Feature Engineering — Create Car Age**

```python
# Get current year
current_year = pd.Timestamp.now().year
```

```python
# Create new column 'car_age'
df['car_age'] = current_year - df['year']
```

```python
# Preview updated dataset
df.head()
```

### **Outlier Detection (price vs km_driven)**

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt

# Scatter plot to visually detect outliers
plt.figure(figsize=(8,6))
plt.scatter(df['km_driven'], df['price'])
plt.xlabel("KM Driven")
plt.ylabel("Price")
plt.title("Outlier Detection: Price vs KM Driven")
plt.scatter(df['km_driven'], df['price'], color='purple')
plt.show()
# Calculate Q1 and Q3
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

#Remove outliers using IQR method
# Calculate IQR
IQR = Q3 - Q1

# Filter dataset to remove extreme price outliers
df = df[(df['price'] >= Q1 - 1.5*IQR) & (df['price'] <= Q3 + 1.5*IQR)]
```

### **Categorical Cleaning (fuel_type, owner_type)**

```python
# Standardize text format for categorical columns
df['fuel_type'] = df['fuel_type'].str.strip().str.title()
df['owner_type'] = df['owner_type'].str.strip().str.title()

# Check unique values after cleaning
df['fuel_type'].unique()
df['owner_type'].unique()
```

### **Basic EDA Visualization**

```python
# Average price by brand
df.groupby('brand')['price'].mean().plot(kind='bar',color='teal')
plt.title("Average Price by Brand")
plt.show()
```

## **Charts**

## Univariate Analysis

Univariate Analysis focuses on examining a single variable at a time to understand its distribution, pattern, and key characteristics. In the Indian automobile dataset, this type of analysis helps identify how individual factors such as **fuel type, brand distribution, owner type, price range, transmission type, and vehicle category** are distributed across the market.

Various visualization techniques such as **bar charts, pie charts, histograms, and box plots** are commonly used in univariate analysis. For example, **bar charts** can display the number of cars available for each brand, while **pie charts** help visualize the percentage distribution of fuel types like petrol, diesel, or CNG. **Histograms** can show the distribution of vehicle prices or mileage ranges, and **box plots** help identify outliers in numerical variables such as price or kilometers driven.

Through this analysis, we can easily observe which categories dominate the dataset. For instance, we can determine which **fuel type is most commonly used, which brands appear most frequently, the distribution of transmission types, or how car prices are spread across the market**. This analysis provides a foundational understanding of the dataset before exploring relationships between multiple variables.


## Bivariate Analysis

Bivariate Analysis examines the relationship between two variables to understand how one variable influences or relates to another. In the context of the Indian automobile dataset, this analysis helps reveal patterns between variables such as **vehicle price and brand, fuel type and mileage, transmission type and price, or kilometers driven and vehicle age**.

Several visualization techniques are used for this analysis, including **scatter plots, grouped bar charts, and box plots**. For example, a **scatter plot** can illustrate the relationship between price and mileage to see whether higher mileage vehicles tend to have lower resale prices. A **grouped bar chart** can compare the average price of vehicles across different brands or fuel types. Similarly, **box plots** can be used to compare price distributions among different transmission types or vehicle categories.

Through bivariate analysis, we can identify meaningful trends such as **which fuel types are associated with higher prices, how transmission type affects vehicle pricing, or how mileage varies across different brands**. This type of analysis provides deeper insights that cannot be observed when variables are analyzed individually.


## Multivariate Analysis

Multivariate Analysis involves examining **three or more variables simultaneously** to understand complex relationships within the dataset. In the Indian automobile analysis, this approach helps explore how multiple factors such as **brand, fuel type, owner type, price, mileage, and transmission** interact and influence the automobile market.

Advanced visualization techniques such as **heatmaps, pair plots, bubble charts, and sunburst charts** are commonly used in multivariate analysis. For instance, a **heatmap** can display the correlation between numerical variables like price, mileage, and kilometers driven. A **pair plot** can show relationships among multiple numerical variables simultaneously. **Bubble charts** allow visualization of three variables at once by using bubble size as an additional dimension, while **sunburst charts** help represent hierarchical relationships such as brand → fuel type → transmission.

Multivariate analysis enables us to uncover **hidden patterns, correlations, and market segments** that are not visible through univariate or bivariate analysis alone. This method provides a deeper and more comprehensive understanding of the Indian automobile market and supports more effective data-driven insights.

### Vehicle Fuel Type Breakdown

```python
import matplotlib.pyplot as plt
import squarify

# Prepare data
fuel_counts = df['Fuel_type'].value_counts()
sizes = fuel_counts.values
total = sizes.sum()

labels = [
    f"{fuel}\n{count} ({count/total:.1%})"
    for fuel, count in zip(fuel_counts.index, sizes)
]

# Better colors
colors = [
    '#1f77b4',  # blue
    '#ff7f0e',  # orange
    '#2ca02c',  # green
    '#d62728'   # red
]

plt.figure(figsize=(12,7))

squarify.plot(
    sizes=sizes,
    label=labels,
    color=colors,
    alpha=0.85,
    pad=True,
    text_kwargs={'fontsize':12,'weight':'bold'}
)

plt.title("Vehicle Fuel Type Breakdown", fontsize=16, fontweight='bold')

plt.axis('off')

plt.text(
    1.02, 0.8,
    '''Insight:

Petrol and diesel vehicles dominate the dataset,
indicating strong dependence on conventional fuels.

Electric and hybrid vehicles represent a smaller share,
highlighting the gradual transition toward cleaner energy
in the automobile market.''',
    transform=plt.gca().transAxes,
    fontsize=11,
    verticalalignment='top'
)

plt.show()
```

### Automobile Vehicle Market Composition

```python
import plotly.express as px

# Prepare hierarchical dataframe
df_hierarchy = df.groupby(['Fuel_type', 'Owner_type', 'Brand']).size().reset_index(name='Count')

# Create Sunburst chart
fig = px.sunburst(
    df_hierarchy,
    path=['Fuel_type', 'Owner_type', 'Brand'],  # hierarchy levels
    values='Count',
    color='Fuel_type',
    title='Automobile Vehicle Market Composition',
    color_discrete_map={
        'Petrol':'#1f77b4',
        'Diesel':'#ff7f0e',
        'Electric':'#2ca02c',
        'Hybrid':'#d62728'
    }
)

# Increase figure size
fig.update_layout(
    width=1000,
    height=800,
    margin=dict(t=50, l=25, r=25, b=80)
)

# Add insight annotation
fig.add_annotation(
    text=(
        "Insight:<br>"
        "- Petrol and diesel dominate the market<br>"
        "- Electric and hybrid vehicles are less frequent<br>"
        "- Owner type and brand distribution visible hierarchically"
    ),
    x=0.5,
    y=-0.15,
    showarrow=False,
    xref='paper',
    yref='paper',
    font=dict(size=14, color="black"),
    align="center"
)

fig.show()
```
